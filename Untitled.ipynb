Data Analysis For Data Science Job Description

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("glassdoor_jobs.csv")
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(
    lambda x: 1 if 'employer' in x.lower() else 0
)
```

```python
df["hourly"].value_counts()
```

```python
df['employer_provided'].value_counts()
```

Salary Analysis

```python
#Clean Salary Data
df = df[df['Salary Estimate'] != '-1']
salary = df['Salary Estimate'].str.replace('K','', regex=False)\
                               .str.replace('$','', regex=False)\
                               .str.replace('Per Hour','', regex=False)\
                               .str.replace('Employer Provided Salary:','', regex=False)\
                               .str.split('(').str[0]
```

```python
#Minimun salary
df['min_salary'] = salary.apply(lambda x: int(x.split('-')[0]))
```

```python
#Extract maximum salary
df['max_salary'] = salary.apply(lambda x: int(x.split('-')[1]))
```

```python
#Avarege Salary
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2
```

Company Name (Text Only)

```python
df['company_text'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
```

State Field

```python
df['state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

Age of Company

```python
df['company_age'] = 2026 - df['Founded']
```

```python
df['company_age'] = df['company_age'].apply(lambda x: x if x > 0 else None)
```
